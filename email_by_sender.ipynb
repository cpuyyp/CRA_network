```python
import pandas as pd
from function_library import *
from function_library2 import *
```

```python
body = pd.read_csv('new_clean_output.csv', index_col=0)['Body'].values.tolist()
```

```python
from_list = fromPickle('from_list')
from_list = standardize_triplet(from_list, tuple)
unique_from_list = list(set(from_list))
unique_from_list = sorted(unique_from_list)
```

```python
all_email_body = []
for i in range(len(unique_from_list)):
    all_email_body.append([])
for i, sender in enumerate(from_list):
    idx = unique_from_list.index(sender)
    all_email_body[idx].append(body[i])
```

```python
df_new = pd.DataFrame(data=list(zip(unique_from_list,all_email_body)),columns=['sender', 'email_bodies'])
df_new.to_csv('emails_by_sender.csv')
```
