```python
def search_emails(year, week=None,day=None):    
    if week==None and day==None:
        raise Exception('Input week or day')
    elif week!=None and day!=None:
        raise Exception('Input week or day, only accept one')
    search_index = []
    for i in range(len(day_list)):
        if year_list[i] < year:
            continue
        if year_list[i] == year:
            if week!=None and week_list[i] == week:
                search_index.append(i)
            elif day!=None and day_list[i] == day:
                search_index.append(i)
        if year_list[i] > year:
            break
        search_result = []
        for i in search_index:
            search_result.append(bodies_in_a_line[i])
    return search_result
```

```python
# https://github.com/CircusMonkey/covariance-ellipsoid/blob/master/ellipsoid.py
def get_cov_ellipsoid(cov, mu=np.zeros((3)), nstd=3):
    assert cov.shape==(3,3)

    # Find and sort eigenvalues to correspond to the covariance matrix
    eigvals, eigvecs = np.linalg.eigh(cov)
    idx = np.sum(cov,axis=0).argsort()
    eigvals_temp = eigvals[idx]
    idx = eigvals_temp.argsort()
    eigvals = eigvals[idx]
    eigvecs = eigvecs[:,idx]

    # Set of all spherical angles to draw our ellipsoid
    n_points = 100
    theta = np.linspace(0, 2*np.pi, n_points)
    phi = np.linspace(0, np.pi, n_points)

    # Width, height and depth of ellipsoid
    rx, ry, rz = nstd * np.sqrt(eigvals)

    # Get the xyz points for plotting
    # Cartesian coordinates that correspond to the spherical angles:
    X = rx * np.outer(np.cos(theta), np.sin(phi))
    Y = ry * np.outer(np.sin(theta), np.sin(phi))
    Z = rz * np.outer(np.ones_like(theta), np.cos(phi))

    # Rotate ellipsoid for off axis alignment
    old_shape = X.shape
    # Flatten to vectorise rotation
    X,Y,Z = X.flatten(), Y.flatten(), Z.flatten()
    X,Y,Z = np.matmul(eigvecs, np.array([X,Y,Z]))
    X,Y,Z = X.reshape(old_shape), Y.reshape(old_shape), Z.reshape(old_shape)
   
    # Add in offsets for the mean
    X = X + mu[0]
    Y = Y + mu[1]
    Z = Z + mu[2]
    
    return X,Y,Z
```

```python
# load file and build LIWC dictionary
with open('words_in_LIWC_dict.txt','r') as f:
    lines = f.readlines()
LIWC_dic = {}
for line in tqdm_notebook(lines):
    line = line.split(',')
    key = line[0]
    category = (np.array(line[1:])== '1').astype(int)
    LIWC_dic[key] = category
```

```python
# get LIWC score for each email 
LIWC_freq_mat = np.zeros((len(bodies_in_a_line), 74))
LIWC_count_mat = np.zeros((len(bodies_in_a_line), 74))

for idx,body in enumerate(tqdm_notebook(bodies_in_a_line)):
    counts = np.zeros(74)
    words = body.lower().replace('.','').replace(',','').replace('\'','').replace('(',' ').replace(')',' ').replace('/',' ').replace('-',' ').split()
    body_length = len(words)
    for word in words:
        try:
            counts = counts + LIWC_dic[word]
        except:
#             print('word',word,'not in LIWC dictionary!')
            pass
    if body_length != 0:
        LIWC_freq_mat[idx,:] = counts/body_length
        LIWC_count_mat[idx,:] = counts 
```

```python
# extract emails
rich_LIWC = LIWC_freq_mat[rick_emails_idx]
ryan_LIWC = LIWC_freq_mat[ryan_emails_idx]
mark_LIWC = LIWC_freq_mat[mark_emails_idx]
adam_LIWC = LIWC_freq_mat[adam_emails_idx]
all_LIWC = np.vstack((rich_LIWC, ryan_LIWC, mark_LIWC, adam_LIWC))
```
